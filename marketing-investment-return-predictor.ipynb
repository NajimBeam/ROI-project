```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df_test = pd.read_csv('test.csv', encoding='ISO-8859-1', sep=";")
df_train = df
```

```python
## Checking for Linearity - The relationship seems to be linear to a great extent
plt.scatter(df['3D ROAS'],df['30D ROAS'])
plt.xlim(0,5000)
plt.ylim(0,10000)
```

```python
## Checking for Linearity - The relationship seems to be linear
plt.scatter(df['7D ROAS'],df['30D ROAS'])
plt.xlim(0,5000)
plt.ylim(0,10000)
```

```python
# These two independent variables are the only "very early" indicators that produce an acceptable p value (< 0.05) and exist 
# in the testing dataset such that they can be used to predict the 30D ROAS values
x = df_train[['3D ROAS','7D ROAS']]

# This is the dependent variable
y = df_train['30D ROAS']
```

```python
# Our visual inspections have supported the notion that our predictor variables are linearly related to our outcome variable
# Thus, I have decided to fit a linear regression model
model = sm.OLS(y,x).fit()
predictions = model.predict(x)
```

```python
model.summary()
```

```python
# Relatively medium-sized R-Squared (only mediocre predictive power)
# Value with all statistical variables significant
```
